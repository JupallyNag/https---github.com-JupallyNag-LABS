```python
import pandas as pd
df1 = pd.read_csv("C:/Users/DELL/Documents/GitHub/IH_DA_FT_MAR_2023/Class_Materials/Case_Studies/Customer_Analysis_Case_Study/Data/file1.csv")
print(df1.columns)
df2 = pd.read_csv("C:/Users/DELL/Documents/GitHub/IH_DA_FT_MAR_2023/Class_Materials/Case_Studies/Customer_Analysis_Case_Study/Data/file2.csv")
print(df2.columns)
df3 = pd.read_csv("C:/Users/DELL/Documents/GitHub/IH_DA_FT_MAR_2023/Class_Materials/Case_Studies/Customer_Analysis_Case_Study/Data/file3.csv")
print(df3.columns)
```

```python
df3.rename(columns={'State': 'ST', 'Gender': 'GENDER'}, inplace=True)
print(df3.columns)
```

```python
df4=pd.concat([df1, df2,df3], axis=0)
display(df4)
```

```python
df4.sort_values("Customer", inplace=True)
# display(df4)
df4.drop_duplicates(keep=False, inplace=True)
display(df4)
```

```python
# df4.info()
df4= df4.fillna(0)
df4['Customer Lifetime Value']=df4['Customer Lifetime Value'].str.replace('%','')
# display(df4)
df4['Customer Lifetime Value'] = df4['Customer Lifetime Value'].apply(pd.to_numeric, errors='coerce')
df4['Total Claim Amount'] = df4['Total Claim Amount'].apply(pd.to_numeric, errors='coerce')
df4['Number of Open Complaints'] = df4['Number of Open Complaints'].apply(pd.to_numeric, errors='coerce')
df4= df4.fillna(0)
df4.info()
```

```python
convert_dict = {'Customer Lifetime Value': int,
                'Total Claim Amount': int,
                'Number of Open Complaints':int
                }
 
df4 = df4.astype(convert_dict)
```
